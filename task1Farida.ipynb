```python
import numpy as np
import cv2
import json
from scipy import ndimage
from PIL import Image, ImageOps 
from skimage import color
import matplotlib.pyplot as plt
```

```python
#Roberts, Prewitt, Sobel 3x3, and Sobel 5x5 operators on "city.jpg"
```

```python
#1.Roberts
imgR = cv2.imread("city.jpg",0).astype('float64')
imgR/=255.0

roberts_cross_v = np.array( [[1, 0 ], [0,-1 ]] )
roberts_cross_h = np.array( [[ 0, 1 ], [ -1, 0 ]] )

vertical = ndimage.convolve( imgR, roberts_cross_v )
horizontal = ndimage.convolve( imgR, roberts_cross_h )

edged_img = np.sqrt( np.square(horizontal) + np.square(vertical))
edged_img*=255

cv2.imwrite("outputRoberts.jpg",edged_img)
robertsImage = cv2.imread("outputRoberts.jpg")
plt.imshow(robertsImage)
```

```python
#Prewitt
imgP=cv2.imread("city.jpg")
imgP=cv2.cvtColor(src=imgP,code =cv2.COLOR_BGR2GRAY)
prewittImage = ndimage.prewitt(imgP)

cv2.imwrite("outputPrewitt.jpg",prewittImage)
prewittImage = cv2.imread("outputPrewitt.jpg")
plt.imshow(prewittImage)
```

```python
#3. Sobel 3x3
imgS3=cv2.imread("city.jpg")

imgB3 = cv2.GaussianBlur(imgS3,(3,3), sigmaX=0, sigmaY=0)
sobel3Image = cv2.Sobel(src=imgB3, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3)

cv2.imwrite("outputSobel3x3.jpg",sobel3Image)
sobel3Image = cv2.imread("outputSobel3x3.jpg")
plt.imshow(sobel3Image)
```

```python
#4. Sobel 5x5
imgS3=cv2.imread("city.jpg")

imgB5 = cv2.GaussianBlur(imgS3,(3,3), sigmaX=0, sigmaY=0)

sobel5Image = cv2.Sobel(src=imgB5, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)

cv2.imwrite("outputSobel5x5.jpg",sobel5Image)
sobel5Image = cv2.imread("outputSobel5x5.jpg")
plt.imshow(sobel5Image)
```

```python
#5. Plotting the previous 4 edge detection results
fig, axs = plt.subplots(2, 2, dpi=300)
fig.tight_layout(pad=2)
axs[0,0].imshow(robertsImage)
axs[0,0].set_title("Roberts")
axs[0,1].imshow(prewittImage)
axs[0,1].set_title("Prewitt")
axs[1,0].imshow(sobel3Image)
axs[1,0].set_title("Sobel 3x3")
axs[1,1].imshow(sobel5Image)
axs[1,1].set_title("Sobel 5x5")
```
